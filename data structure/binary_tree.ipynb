```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def to_binary_tree(items) -> TreeNode:
    """Create BT from list of values."""
    n = len(items)
    # Edge Case: n = 0
    if n == 0:
        return None

    # inner的input是items[index]的數值當做TreeNode的val
    def inner(index: int = 0) -> TreeNode:
        """Closure function using recursion to build tree."""
        if n <= index or items[index] is None:
            return None
        # 用每一個list element當作TreeNode的val
        node = TreeNode(items[index])
        # node下的left/right是
        node.left = inner(2 * index + 1)
        node.right = inner(2 * index + 2)
        return node

    return inner()
```

```python
root = [4,2,7,1,3,6,9]
bt = to_binary_tree(root)
```

get any level of tree value

```python
def get_value_in_binary_tree(bt, traversal):
    if len(traversal) == 0:
        print('bt.val:', bt.val)
        print('bt.left:', bt.left)
        print('bt.right:', bt.right)
        return
    
    traversal_lr = traversal.pop(0)
    if traversal_lr == 'l':
        get_value_in_binary_tree(bt.left, traversal)
    if traversal_lr == 'r':
        get_value_in_binary_tree(bt.right, traversal)
```

```python
get_value_in_binary_tree(bt, [])
```

```python
get_value_in_binary_tree(bt, ['l'])
```

```python
get_value_in_binary_tree(bt, ['l','r'])
```

```python
get_value_in_binary_tree(bt, ['l','l'])
```
