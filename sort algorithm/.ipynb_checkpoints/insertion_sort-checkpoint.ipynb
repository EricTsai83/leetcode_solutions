```python
nums = [4,2,7,1,3]

def insertion_sort(nums):
    '''
    從第二個元素開始當tmp_val，在每個階段把比tmp_val大的數值往右
    '''
    for i in range(1,len(nums)):
        tmp_val = nums[i]
        print(tmp_val)
        tmp_pos = i

        # 遍歷當前tmp_val以前的數組, 把比tmp_val大的值往右靠
        
        while tmp_pos > 0 and nums[tmp_pos-1] > tmp_val:
            print(nums)
            nums[tmp_pos] = nums[tmp_pos -1]
            tmp_pos -= 1
            print(nums)
        print('*End of while loop*')

        # 最後停住的指針, 再把tmp_val插入
        nums[tmp_pos] = tmp_val
        print(nums)
        print('----------End of iter----------')
        
    return nums

insertion_sort(nums)
```
