```python
nums = [4,2,7,1,3]

def insertion_sort(nums):
    '''
    tmp_valtmp_val
    '''
    for i in range(1,len(nums)):
        tmp_val = nums[i]
        print(tmp_val)
        tmp_pos = i

        # tmp_val, tmp_val
        
        while tmp_pos > 0 and nums[tmp_pos-1] > tmp_val:
            print(nums)
            nums[tmp_pos] = nums[tmp_pos -1]
            tmp_pos -= 1
            print(nums)
        print('*End of while loop*')

        # , tmp_val
        nums[tmp_pos] = tmp_val
        print(nums)
        print('----------End of iter----------')
        
    return nums

insertion_sort(nums)
```
